```python
import ROOT #Start by importing ROOT
import numpy as np

# ROOT.enableJSVis() #need this to make interactive ROOT plots
c1 = ROOT.TCanvas() #declare the ROOT canvas


ROOT.gSystem.Load("/usr/local/software/quadis/build/lib/libquadis.so")
```

```python
infile = ROOT.TFile("/home/vasu/quadis/old_build/bin/base/noise_1.root")
tree = infile.Get("T2")
num_files = tree.GetEntries()
print (num_files)
```

```python
h1 = ROOT.TH1F("h1","h1;MaxAbs;Counts",1000,-15,15)
tree.Draw("trapPulse_MaxAbs>>h1","(-15<trapPulse_MaxAbs) && (trapPulse_MaxAbs<15)")
c1.Draw()
c1.SetWindowSize(1000,1000)
c1.SetCanvasSize(700,500)
c1.SetLogy()
h1.GetXaxis().SetLabelSize(0.06)
h1.GetYaxis().SetLabelSize(0.06)
h1.GetXaxis().SetTitleSize(444)
h1.GetYaxis().SetTitleSize(444)
h1.SetLineWidth(2)
h1.SetStats(0)
```

```python
h2 = ROOT.TH1F("h2","h2;Baseline;Counts",1000,3.0e4,3.5e4)
tree.Draw("blrPulse_Baseline>>h2","(30000>trapPulse_MaxAbs) && (trapPulse_MaxAbs<35000)")
c1.Draw()
c1.SetWindowSize(1000,1000)
c1.SetCanvasSize(700,500)
c1.SetLogy()
h2.GetXaxis().SetLabelSize(0.06)
h2.GetYaxis().SetLabelSize(0.06)
h2.GetXaxis().SetTitleSize(444)
h2.GetYaxis().SetTitleSize(100)
h2.GetYaxis().SetMaxDigits(2)
h2.GetXaxis().SetMaxDigits(3)
h2.SetLineWidth(2)
h2.SetStats(0)
```
