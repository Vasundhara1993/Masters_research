```python
import ROOT #Start by importing ROOT
import numpy as np
ROOT.enableJSVis() #need this to make interactive ROOT plots
import os
import time
c1 = ROOT.TCanvas() #declare the ROOT canvas
legend = ROOT.TLegend()
#ROOT.gSystem.Load("/home/vasu/quadis/build/lib/libquadis.so")
ROOT.gSystem.Load("/usr/local/software/quadis/build/lib/libquadis.so") 
```

```python
infile = ROOT.TFile("/home/vasu/PPC/PPC_SigGen/SCAN100004/SCAN100004_RUN13.root")
tree = infile.Get("T1")
nevent = ROOT.NEvent()
tree.SetBranchAddress("NEvent",nevent)
tree.GetEntry(5)
npulse = nevent.GetPulse(0) #the original pulse

#Initialize the fitter1
fitter1 = ROOT.NAPulseFitter("/home/vasu/PPC/PPC_SigGen/SignalLib_root/new_OutputRZ.root","T1","NEvent",1)
baselines = 1000
taus = 7100
rises=750
gaps=225
pulse = fitter1.GetPreProcessedPulse(npulse, False, baselines, taus, rises, gaps)
fitter1.SetLibPulseMinThreshold(0.01)
fitter1.SetLibPulseMaxThreshold(0.95)
fitter1.SetShiftWindow(5) 
fitter1.FitPulse(pulse)
fitter1.PrintBestFit()
libpulse = fitter1.GetLibPulse(724)
shift = fitter1.GetBestShift()
lh1 = libpulse.GetTH1Dmus()
lh1.SetLineColor(1)
lh1.SetLineWidth(2)
# lh1.GetXaxis().SetLimits(8*shift/1000.,8*shift/1000.+lh1.GetNbinsX()*8/1000.)
lh1.Draw()

#Initialize the fitter2
fitter2 = ROOT.NAPulseFitter("/home/vasu/PPC/PPC_SigGen/SignalLib_root/new_OutputRZAlpha1.root","T1","NEvent",1)
baselines = 1000
taus = 7100
rises=750
gaps=225
pulse = fitter2.GetPreProcessedPulse(npulse, False, baselines, taus, rises, gaps)
fitter2.SetLibPulseMinThreshold(0.01)
fitter2.SetLibPulseMaxThreshold(0.95)
fitter2.SetShiftWindow(5) 
fitter2.FitPulse(pulse)
fitter2.PrintBestFit()
libpulse = fitter2.GetLibPulse(724)
shift = fitter2.GetBestShift()
lh2 = libpulse.GetTH1Dmus()
lh2.SetLineColor(2)
lh2.SetLineWidth(2)
# lh2.GetXaxis().SetLimits(8*shift/1000.,8*shift/1000.+lh2.GetNbinsX()*8/1000.)
lh2.Draw("Same")

#Initialize the fitter3
fitter3 = ROOT.NAPulseFitter("/home/vasu/PPC/PPC_SigGen/SignalLib_root/new_OutputRZAlpha2.root","T1","NEvent",1)
baselines = 1000
taus = 7100
rises=750
gaps=225
pulse = fitter3.GetPreProcessedPulse(npulse, False, baselines, taus, rises, gaps)
fitter3.SetLibPulseMinThreshold(0.01)
fitter3.SetLibPulseMaxThreshold(0.95)
fitter3.SetShiftWindow(5) 
fitter3.FitPulse(pulse)
fitter3.PrintBestFit()
libpulse = fitter3.GetLibPulse(724)
shift = fitter3.GetBestShift()
lh3 = libpulse.GetTH1Dmus()
lh3.SetLineColor(3)
lh3.SetLineWidth(2)
# lh3.GetXaxis().SetLimits(8*shift/1000.,8*shift/1000.+lh3.GetNbinsX()*8/1000.)
lh3.Draw("Same")

# Initialize the fitter4
fitter4 = ROOT.NAPulseFitter("/home/vasu/PPC/PPC_SigGen/SignalLib_root/new_OutputRZAlpha3.root","T1","NEvent",1)
baselines = 1000
taus = 7100
rises=750
gaps=225
pulse = fitter4.GetPreProcessedPulse(npulse, False, baselines, taus, rises, gaps)
fitter4.SetLibPulseMinThreshold(0.01)
fitter4.SetLibPulseMaxThreshold(0.95)
fitter4.SetShiftWindow(5) 
fitter4.FitPulse(pulse)
fitter4.PrintBestFit()
libpulse = fitter4.GetLibPulse(724)
shift = fitter4.GetBestShift()
lh4 = libpulse.GetTH1Dmus()
lh4.SetLineColor(4)
lh4.SetLineWidth(2)
# lh4.GetXaxis().SetLimits(8*shift/1000.,8*shift/1000.+lh4.GetNbinsX()*8/1000.)
lh4.Draw("Same")

# Initialize the fitter5
fitter5 = ROOT.NAPulseFitter("/home/vasu/PPC/PPC_SigGen/SignalLib_root/new_OutputRZAlpha4.root","T1","NEvent",1)
baselines = 1000
taus = 7100
rises=750
gaps=225
pulse = fitter5.GetPreProcessedPulse(npulse, False, baselines, taus, rises, gaps)
fitter5.SetLibPulseMinThreshold(0.01)
fitter5.SetLibPulseMaxThreshold(0.95)
fitter5.SetShiftWindow(5) 
fitter5.FitPulse(pulse)
fitter5.PrintBestFit()
libpulse = fitter5.GetLibPulse(724)
shift = fitter5.GetBestShift()
lh5 = libpulse.GetTH1Dmus()
lh5.SetLineColor(5)
lh5.SetLineWidth(2)
# lh4.GetXaxis().SetLimits(8*shift/1000.,8*shift/1000.+lh4.GetNbinsX()*8/1000.)
lh5.Draw("Same")

# Initialize the fitter6
fitter6 = ROOT.NAPulseFitter("/home/vasu/PPC/PPC_SigGen/SignalLib_root/new_OutputRZAlpha5.root","T1","NEvent",1)
baselines = 1000
taus = 7100
rises=750
gaps=225
pulse = fitter6.GetPreProcessedPulse(npulse, False, baselines, taus, rises, gaps)
fitter6.SetLibPulseMinThreshold(0.01)
fitter6.SetLibPulseMaxThreshold(0.95)
fitter6.SetShiftWindow(5) 
fitter6.FitPulse(pulse)
fitter6.PrintBestFit()
libpulse = fitter6.GetLibPulse(724)
shift = fitter6.GetBestShift()
lh6 = libpulse.GetTH1Dmus()
lh6.SetLineColor(6)
lh6.SetLineWidth(2)
# lh4.GetXaxis().SetLimits(8*shift/1000.,8*shift/1000.+lh4.GetNbinsX()*8/1000.)
lh6.Draw("Same")

# Initialize the fitter7
fitter7 = ROOT.NAPulseFitter("/home/vasu/PPC/PPC_SigGen/SignalLib_root/new_OutputRZAlpha6.root","T1","NEvent",1)
baselines = 1000
taus = 7100
rises=750
gaps=225
pulse = fitter7.GetPreProcessedPulse(npulse, False, baselines, taus, rises, gaps)
fitter7.SetLibPulseMinThreshold(0.01)
fitter7.SetLibPulseMaxThreshold(0.95)
fitter7.SetShiftWindow(5) 
fitter7.FitPulse(pulse)
fitter7.PrintBestFit()
libpulse = fitter7.GetLibPulse(724)
shift = fitter7.GetBestShift()
lh7 = libpulse.GetTH1Dmus()
lh7.SetLineColor(7)
lh7.SetLineWidth(2)
# lh4.GetXaxis().SetLimits(8*shift/1000.,8*shift/1000.+lh4.GetNbinsX()*8/1000.)
lh7.Draw("Same")



legend.SetHeader("Legend","C")
legend.AddEntry(lh2,"Alpha = 0.053","l")
legend.AddEntry(lh1,"Alpha = 0.071","l")
legend.AddEntry(lh3,"Alpha = 0.0901","l")
legend.AddEntry(lh4,"Alpha = 0.135","l")
legend.AddEntry(lh5,"Alpha = 0.188","l")
legend.AddEntry(lh6,"Alpha = 0.308","l")
legend.AddEntry(lh7,"Alpha = 0.347","l")
legend.Draw()

c1.Draw()
# c1.SetWindowSize(1000,1000)
# c1.SetCanvasSize(700,500)
lh1.GetXaxis().SetRange(0,105)
lh1.GetXaxis().SetLabelSize(0.06)
lh1.GetYaxis().SetLabelSize(0.06)
lh1.GetXaxis().SetTitleSize(444)
lh1.GetYaxis().SetTitleSize(444)
lh1.SetStats(0)
```
