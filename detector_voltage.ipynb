```python
import numpy as np
import matplotlib.pyplot as plt
import ROOT
import os
import time
ROOT.enableJSVis()
c1 = ROOT.TCanvas()
ROOT.gSystem.Load("/home/vasu/quadis/build/lib/libquadis.so")
import pandas as pd
```

```python
df = pd.read_csv('/home/vasu/PPC/PPC_SigGen/voltage/ev3500.dat',header = None,delim_whitespace = True,skiprows = 36)
```

```python
num = df.shape[0]
```

```python
hv = ROOT.TH2F("hv","hv",100,0,33,100,0,51)
hx = hv.GetXaxis().GetBinWidth(1)
hy = hv.GetYaxis().GetBinWidth(1)

for i in range (num):
    hv.SetBinContent(int(df[0][i]/hx),int(df[1][i]/hy),df[2][i])
```

```python
hv.Draw("colz")
c1.SetWindowSize(500,510);
c1.SetCanvasSize(330,510);
hv.GetXaxis().SetLabelSize(0.06)
hv.GetYaxis().SetLabelSize(0.06)
hv.GetZaxis().SetLabelSize(0.05)
hv.GetXaxis().SetTitleSize(444)
hv.GetYaxis().SetTitleSize(444)
hv.SetStats(0)
# c1.SaveAs("images/3500V.png")
c1.SetRightMargin(0.2)
c1.Draw()
```
