```python
data_ind = []
num_files=0
f = open('rt_data_ind', 'r')
while True:
    line = f.readline()
    if not line:
        break
    else:
        temp = line.strip("\n")
        data_ind.append(int(temp))
    num_files = num_files+1
f.close()
infile = ROOT.TFile("/home/vasu/PPC/PPC_SigGen/SCAN100004/SCAN100004_RUN13.root")
tree = infile.Get("T1")
nevent = ROOT.NEvent()
tree.SetBranchAddress("NEvent",nevent)
data = np.zeros(4096)
x = np.arange(0,4096)
for i in range (num_files):
    j = data_ind[i]
    print j
    tree.GetEntry(j)
    npulse = nevent.GetPulse(0)
    for k in range (len(npulse)):
        data[k] = npulse[k]
    plt.plot(x,data)
    plt.show()
```

```python
infile = ROOT.TFile("/home/vasu/PPC/PPC_SigGen/SCAN100004/SCAN100004_RUN13.root")
tree = infile.Get("T1")
nevent = ROOT.NEvent()
tree.SetBranchAddress("NEvent",nevent)
tree.GetEntry(5)
npulse = nevent.GetPulse(0) #the original pulse

fitter = ROOT.NAPulseFitter("/home/vasu/PPC/PPC_SigGen/SignalLib_root/new_OutputRZperfect.root","T1","NEvent",1)
baselines = 1000
taus = 7100
rises=750
gaps=225
pulse = fitter.GetPreProcessedPulse(npulse, False, baselines, taus, rises, gaps)
fitter.SetLibPulseMinThreshold(0.01)
fitter.SetLibPulseMaxThreshold(0.95)
fitter.SetShiftWindow(5) 
fitter.FitPulse(pulse)
fitter.PrintBestFit()
libpulse = fitter.GetBestFitPulse()
shift = fitter.GetBestShift()
lh = libpulse.GetTH1Dmus()
lh.SetLineColor(2)
lh.SetLineWidth(3)
ph = pulse.GetTH1Dmus()
ph.Draw()
lh.GetXaxis().SetLimits(8*shift/1000.,8*shift/1000.+lh.GetNbinsX()*8/1000.)
lh.Draw("same")
c1.Draw()
```
