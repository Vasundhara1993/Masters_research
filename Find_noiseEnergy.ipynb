```python
import numpy as np
import matplotlib.pyplot as plt
import ROOT
import os
ROOT.enableJSVis()
c1 = ROOT.TCanvas()
ROOT.gSystem.Load("/usr/local/software/quadis/build/lib/libquadis.so")   
```

```python
infile = ROOT.TFile("/home/vasu/PPC/PPC_SigGen/NoisyPulses/dataform_RUN6_100.root")
tree = infile.Get("T1")
nevent = ROOT.NEvent()
tree.SetBranchAddress("NEvent",nevent)
tree.GetEntry(0)

npulse = nevent.GetPulse(5) #the original pulse
sP = npulse.GetSamplingPeriod()

blrPulse = ROOT.NPulse()
blrFilter = ROOT.NFBaselineRemover(1000*sP) 
blrFilter.Filter(npulse,blrPulse)
baseline = blrFilter.GetCalcParameterByName("Baseline")
baselineRMS = blrFilter.GetCalcParameterByName("BaselineRMS")
print("the baseline of the pulse was",baselineRMS)


trapPulse = ROOT.NPulse() 
trapFilter = ROOT.NFTrapezoid(200*sP,400*sP,6750*sP,baseline) 
trapFilter.Filter(npulse,trapPulse)
print("The maximum of the trapezoid is",trapFilter.GetCalcParameterByName("Max"))

blrh = blrPulse.GetTH1Dmus()
traph = trapPulse.GetTH1Dmus()
traph.SetLineColor(2)#make this one red

blrh.Draw()
traph.Draw("Same")
c1.Draw()
```

```python
infile = ROOT.TFile("/home/vasu/PPC/PPC_SigGen/NoisyPulses/dataform_RUN6_100.root")
tree = infile.Get("T1")
nevent = ROOT.NEvent()
tree.SetBranchAddress("NEvent",nevent)
nf = tree.GetEntries()
rms = np.zeros(nf)
trap = np.zeros(nf)

for i in range (nf):
    tree.GetEntry(i)
    npulse = nevent.GetPulse(5)
    sP = npulse.GetSamplingPeriod()
    blrPulse = ROOT.NPulse()
    blrFilter = ROOT.NFBaselineRemover(1000*sP) 
    blrFilter.Filter(npulse,blrPulse)
    baseline = blrFilter.GetCalcParameterByName("Baseline")
    baselineRMS = blrFilter.GetCalcParameterByName("BaselineRMS")
    rms[i] = baselineRMS
    
    trapPulse = ROOT.NPulse() 
    trapFilter = ROOT.NFTrapezoid(200*sP,400*sP,6750*sP,baseline) 
    trapFilter.Filter(npulse,trapPulse)
    maxtrap = trapFilter.GetCalcParameterByName("Max")
    trap[i] = maxtrap

print np.average(rms)
print np.average(trap)

```
